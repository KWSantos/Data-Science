```python
import pandas as pd
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, Input
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.metrics import mean_absolute_error
import sklearn
import numpy as np
```

```python
dataset = pd.read_csv('vgsales.csv')
dataset
```

```python
dataset =  dataset.drop('Other_Sales', axis=1)
dataset =  dataset.drop('Global_Sales', axis=1)
dataset
```

```python
dataset.isna().sum()
```

```python
dataset = dataset.dropna()
dataset
```

```python
dataset.shape
```

```python
dataset['Name'].value_counts()
```

```python
dataset = dataset.drop('Name', axis=1)
dataset
```

```python
X = dataset.drop(['NA_Sales', 'EU_Sales', 'JP_Sales'], axis=1).values
Y_na = dataset['NA_Sales'].values
Y_eu = dataset['EU_Sales'].values
Y_jp = dataset['JP_Sales'].values
```

```python
X.shape
```

```python
oneHotEncoder = ColumnTransformer(transformers=[('OneHot', OneHotEncoder(), [1, 3, 4])], remainder='passthrough')
X = oneHotEncoder.fit_transform(X)
```

```python
X.shape
```

```python
input_layer = Input(shape=(X.shape[1],))
hidden_layer1 = Dense(128, activation='relu')(input_layer)
hidden_layer2 = Dense(128, activation='relu')(hidden_layer1)
output_layer1 = Dense(1, activation='linear')(hidden_layer2)
output_layer2 = Dense(1, activation='linear')(hidden_layer2)
output_layer3 = Dense(1, activation='linear')(hidden_layer2)
```

```python
regressor = Model(inputs=input_layer, outputs=[output_layer1, output_layer2, output_layer3])
```

```python
regressor.compile(optimizer='adam', loss='mean_squared_error')
```

```python
X = X.astype(np.float64)
Y_na = Y_na.astype(np.float64)
Y_eu = Y_eu.astype(np.float64)
Y_jp = Y_jp.astype(np.float64)
regressor.fit(X, [Y_na, Y_eu, Y_jp], epochs=500, batch_size=100)
```

```python
predict_na, predict_eu, predict_jp = regressor.predict(X)
```

```python
predict_na.mean(), Y_na.mean()
```

```python
predict_eu.mean(), Y_eu.mean()
```

```python
predict_jp.mean(), Y_jp.mean()
```

```python
mean_absolute_error(Y_na, predict_na)
```

```python
mean_absolute_error(Y_eu, predict_eu)
```

```python
mean_absolute_error(Y_jp, predict_jp)
```
